```python
import pandas as pd
import numpy as np

# 1. Calulating mean population, sum population, max population , min population, mean GDP, standard deviation population GDP, sum GDP


df = pd.read_csv('summary_statistics.csv') 


print(df.head())


print("Pandas Summary Statistics:")

print(f"Mean Population: {df['Population'].mean()}")
print(f"Sum Population: {df['Population'].sum()}")
print(f"Max Population: {df['Population'].max()}")
print(f"Min Population: {df['Population'].min()}")

print(f"Mean GDP: {df['GDP'].mean()}")
print(f"Sum GDP: {df['GDP'].sum()}")
print(f"Max GDP: {df['GDP'].max()}")
print(f"Min GDP: {df['GDP'].min()}")


print("\nNumPy Summary Statistics:")

pop = df['Population'].to_numpy()  
gdp = df['GDP'].to_numpy()

print(f"Mean Population: {np.mean(pop)}")
print(f"Standard Deviation Population: {np.std(pop)}")
print(f"Sum Population: {np.sum(pop)}")

print(f"Mean GDP: {np.mean(gdp)}")
print(f"Standard Deviation GDP: {np.std(gdp)}")
print(f"Sum GDP: {np.sum(gdp)}")



# 2. Increase the GDP column by 10% using vectorized operation
df['GDP'] = df['GDP'] * 1.10  # Increase GDP by 10%
print("GDP increased by 10% (first 5 rows):")
print(df[['Country', 'GDP']].head())


# 3. Create a MultiIndex with Country and Year (Assuming 'Year' column exists)

if 'Year' not in df.columns:
  
    df['Year'] = [2020, 2020, 2020, 2020] 


df.set_index(['Country', 'Year'], inplace=True)
print("\nDataFrame with MultiIndex (Country, Year):")
print(df.head())


df_swapped = df.swaplevel('Country', 'Year')
print("\nMultiIndex levels swapped (Year, Country):")
print(df_swapped.head())


df_swapped_sorted = df_swapped.sort_index(level=0)
print("\nDataFrame sorted by Year:")
print(df_swapped_sorted.head())


# 4. Use unstack() to pivot Year to columns, Country stays as row index

df_unstacked = df.unstack(level='Year')
print("\nDataFrame after unstacking Year to columns:")
print(df_unstacked.head())

df_unstacked.columns = ['{}_{}'.format(col[0], col[1]) for col in df_unstacked.columns]
print("\nDataFrame with flattened columns after unstack:")
print(df_unstacked.head())


# 5. Hierarchical indexing to organize dataset by Country and Year is already done by set_index above.

country_name = 'USA'
population_trend = df.loc[country_name]['Population']
print(f"\nPopulation trend for {country_name}:")
print(population_trend)


```
